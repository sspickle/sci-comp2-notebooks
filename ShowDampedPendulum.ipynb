```python
import matplotlib.pyplot as plt
import numpy as np

def calculate( theta, omega, t, length, dt, q, Fd, fd):
    # 
    # calculate 10 periods of oscillation .. 
    #

    while 1:
        t.append(t[-1] + dt)
        omega.append( omega[-1] - (g/length)*np.sin(theta[-1])*dt - \
                      q*omega[-1]*dt + Fd*np.sin( fd * t[-1] ) *dt )
        theta.append(theta[-1] + omega[-1]*dt)

        if theta[-1] > np.pi:
            theta[-1] = theta[-1] - 2*np.pi
        if theta[-1] < -np.pi:
            theta[-1] = theta[-1] + 2*np.pi

        if t[-1] > 10*period:
            break

t = [0.0]
omega = [0.0]
theta = [0.2]

length = 9.8
dt = 0.04 
q = 0.5
fd = 0.6667
Fd = 0.0

g = 9.8
period = 2*np.pi/(np.sqrt(g/length))

calculate( theta, omega, t, length, dt, q, Fd, fd )

plt.figure()
plt.plot(t, theta)
plt.xlabel('time')
plt.ylabel(r'$\theta$')
plt.grid()

plt.figure()
plt.plot(theta, omega)
plt.xlabel(r'$\theta$')
plt.ylabel('$\omega$')
plt.grid()

```
