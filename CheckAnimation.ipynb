# Demo of animation

This is one of many ways to animate functions in Jupyter Notebooks.


```python
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.animation as ani
import numpy as np

t = np.linspace(0,2*np.pi)
x = np.sin(t)

fig, ax = plt.subplots()
l, = ax.plot([0,2*np.pi],[-1,1]) # note that this initial plot is just to establish the frame size.

animate = lambda i: l.set_data(t[:i], x[:i])

_ = ani.FuncAnimation(fig, animate, frames=len(t), interval=10, repeat=False)

plt.show()
```
